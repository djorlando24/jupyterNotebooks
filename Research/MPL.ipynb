# Testing MATPLOTLIB functionality in Jupyter

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl

fig=plt.figure()
x = np.linspace(0,10,100)
y = np.sin(x)
ax=fig.add_subplot(111)
ax.scatter(x,y +0.25*(np.random.rand(x.shape[0])-0.5),c='r');
ax.plot(x,y,lw=3);
plt.grid();
plt.title("My Plot");
```
