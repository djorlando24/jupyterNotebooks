```python
# core functions for image handling
import numpy as np
import os, imageio, gzip, time

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib widget
```

```python
# Processed images to load

src = '/Volumes/DUKE SSD ext4/SanjivG_0.5mm_cavitation_imaging/nz_0__5_0_3/saz_20241217_far/nz_0__5_3_co2_00_Re_1__5E4_K_1__12/nz_0__5_3_co2_00_Re_1__5E4_K_1__12_bkRmv.npz.gz'
#src = '/Volumes/DUKE SSD ext4/SanjivG_0.5mm_cavitation_imaging/nz_0__5_0_3/saz_20241217_far/nz_0__5_3_co2_00_Re_2__4E4_K_1__01/nz_0__5_3_co2_00_Re_2__4E4_K_1__01_bkRmv.npz.gz'
```

```python
t0 = time.time()
with gzip.open(src) as gz:
    with np.load(gz) as srcdata:
        bkRmv = srcdata['bkRmv']
        sumIntens=srcdata['sumIntens']
        sumIntens2=srcdata['sumIntens2']
        
print("Read %s array in %.1f min." % (str(bkRmv.shape),(time.time()-t0)/60.))
```

```python
# Stats

meanField=np.nanmean(bkRmv[500:-500,...],axis=0)
stdField=np.nanstd(bkRmv[500:-500,...],axis=0)
```

```python
fig=plt.figure()
plt.imshow(meanField,cmap='gnuplot',vmin=0)
plt.colorbar();
```

```python
# Save movie
t1 = time.time()
output_prefix = src.replace('.npz.gz','')
#bkRmv[bkRmv>4294967295]=4294967294 # for floats

imageio.mimsave(output_prefix+'_video.tif', bkRmv[:1000,...].astype(np.uint16))#, fps=30)
print("Wrote %.0f MB in %.1f min" % (os.path.getsize(output_prefix+'_video.mp4')//1024//1024,\
                                     (time.time()-t1)/60.)))
```
