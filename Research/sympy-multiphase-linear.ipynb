## Problem definition

- Axisymmetric flow in a pipe, nozzle or spray.
- Cylindrical coordinate system.
- Homogeneous in the $\theta$ direction, inhomogeneous in $r$ and $z$ (2D problem).
- Flow is subject to non-modal growth of small perturbations caused by separation of the cavitation from the wall due to REJ or surface irregularities. 
- Multiphase flow, thus compressible and with spatially varying viscosity as well. This adds some complexity on top of typical multiphase flows of this kind such as jets and compressible pipe flow.
- Multiphase flow implies the presence of surface tension forces at gas-liquid interfaces as well, these can be considered as body forces which appear in the nonlinear terms. They are nonlinear because they depend on the local interface curvature which will be a nonlinear function of the local flow variables.
- Base flow can be obtained from CFD simulation , however some post-filtering may be required to obtain an exact steady-state solution to the Navier-Stokes equations (Rolandi et al, 2004).
- Forcing can be determined from experimental data (frequency, mangitude and duty). CFD can get us a mean flow field in a perfect geometry but does not do well at predicting the forcing because it is a consequence of small imperfections and weak feedback loops that occurr irregularly.

## Define variables

- coordinates $r \in [0,R]$, $z \in [-\infty, \infty]$
- fields $\mathbf{u}, p, \rho, x, y, T$ which all are functions of $r$ and $z$. $x$ is the vapour mass fraction and $y$ is the noncondensible phase mass fraction (not coordinates).
- Fields for $\mu, \overline{a}$ can be derived explicitly as functions of the above (mainly, $x,y,T$)
- state variable $\mathbf{q} = \left[ \rho \; \rho u_r \; \rho u_\theta \; \rho u_z \; p \; x \; y \; T \right]^T$
